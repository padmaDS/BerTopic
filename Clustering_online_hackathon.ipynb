Step 6:Make Prediction

```python
#convert to list
new_docs = df.remarks_text.to_list()
```

```python
# Convert floats to strings, keeping other elements unchanged
new_docs = [str(doc) if isinstance(doc, float) else doc for doc in new_docs]
```

```python
topics, probs = model.transform(new_docs)
```

```python
# Get information about topics, including topic names
topic_info = model.get_topic_info()

# Print the topic information
print(topic_info)
```

```python
for doc, topic in zip(new_docs[:10], topics[:10]):
    print("Document:", doc)
    print("Predicted Topic:", topic)
    print()
```

```python
# Map topic IDs to their corresponding names
topic_id_to_name = dict(zip(topic_info['Topic'], topic_info['Name']))

# Print the names of the topics for the first 10 documents
for doc, topic_id in zip(new_docs[10:50], topics[0:20]):
    # Get the name of the topic using the topic ID
    topic_name = topic_id_to_name.get(topic_id, "Unknown Topic")

    print("Document:", doc)
    print("Predicted Topic:", topic_name)
    print()
```

# Complete code at one place

```python
import pandas as pd
import numpy as np
from bertopic import BERTopic
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/no_pii_grievance_v2.csv")

df.head()
# select only 6000 tweets
df = df[0:6000]
model = BERTopic(nr_topics=20, verbose = True)
docs = df.subject_content_text.to_list()
topics, probabilities = model.fit_transform(docs)

#convert to list
new_docs = df.remarks_text.to_list()

# Convert floats to strings, keeping other elements unchanged
new_docs = [str(doc) if isinstance(doc, float) else doc for doc in new_docs]

## :Make Prediction
topics, probs = model.transform(new_docs)

model.save("my_topics_model")
```
